# Model Implementation

```python
data=data.iloc[:,[0,1,2,3,4,6,7,8,9,10,11,12,5]]
```

```python
data.info()
```

```python
X=data.iloc[:,:-1]
```

```python
X
```

```python
Y=data.iloc[:,-1]
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

## Ridge Regressor

```python
from sklearn.linear_model import Ridge
ridge_regressor=Ridge(alpha=0.5)
```

```python
ridge_regressor.fit(x_train,y_train)
```

```python
y_test_predict_ridge=ridge_regressor.predict(x_test)
```

```python
y_train_predict_ridge=ridge_regressor.predict(x_train)
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor=RandomForestRegressor(n_estimators=350,max_depth=8,random_state=50)
```

```python
rf_regressor.fit(x_train,y_train)
```

```python
y_test_predict=rf_regressor.predict(x_test)
```

```python
y_train_predict=rf_regressor.predict(x_train)
```

# Model Evaluation

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from math import sqrt
```

## Ridge Regressor

```python
print("          Training Data Evaluation      ")
print("Mean Absolute error : ",mean_absolute_error(y_train_predict_ridge,y_train))
print("Mean Squared error : ",mean_squared_error(y_train_predict_ridge,y_train))
print("Root Mean Squared error : ",sqrt(mean_squared_error(y_train,y_train_predict_ridge)))
print("R2 Score : ",r2_score(y_train_predict_ridge,y_train))
print("\n")
print("          Testing Data Evaluation      ")
print("Mean Absolute error :  ",mean_absolute_error(y_test_predict_ridge,y_test))
print("Mean Squared error : ",mean_squared_error(y_test_predict_ridge,y_test))
print("Root Mean Squared error : ",sqrt(mean_squared_error(y_test,y_test_predict_ridge)))
print("R2 Score : ",r2_score(y_test,y_test_predict_ridge))
```

## Random Forest Regressor

```python
print("          Training Data Evaluation      ")
print("Mean Absolute error :  ",mean_absolute_error(y_train_predict,y_train))
print("Mean Squared error : ",mean_squared_error(y_train_predict,y_train))
print("Root Mean Squared error : ",sqrt(mean_squared_error(y_train,y_train_predict)))
print("R2 Score : ",r2_score(y_train_predict,y_train))

print("\n")
print("          Testing Data Evaluation      ")
print("Mean Absolute error :  ",mean_absolute_error(y_test_predict,y_test))
print("Mean Squared error : ",mean_squared_error(y_test_predict,y_test))
print("Root Mean Squared error : ",sqrt(mean_squared_error(y_test,y_test_predict)))
print("R2 Score : ",r2_score(y_test,y_test_predict))
```

```python
Models=['Ridge Regressor','Random Forest Regressor']
MAE=[mean_absolute_error(y_train_predict_ridge,y_train),
     mean_absolute_error(y_train_predict,y_train)]

MSE=[mean_squared_error(y_train_predict_ridge,y_train),
     mean_squared_error(y_train_predict,y_train)]


RMSE=[sqrt(mean_squared_error(y_train,y_train_predict_ridge)),
      sqrt(mean_squared_error(y_train,y_train_predict))]

R2_score=[r2_score(y_train_predict_ridge,y_train),
          r2_score(y_train_predict,y_train)]

models_list=list(zip(Models,MAE,MSE,RMSE,R2_score))

print("Training Dataset Outcomes")
models_training=pd.DataFrame(models_list,columns=['Models','MAE','MSE','RMSE','R2_score'])
models_training
```

```python
MAE=[mean_absolute_error(y_test_predict_ridge,y_test),
     mean_absolute_error(y_test_predict,y_test)]


MSE=[mean_squared_error(y_test_predict_ridge,y_test),
     mean_squared_error(y_test_predict,y_test)]


RMSE=[sqrt(mean_squared_error(y_test,y_test_predict_ridge)),
      sqrt(mean_squared_error(y_test,y_test_predict))]

R2_score=[r2_score(y_test,y_test_predict_ridge),
         r2_score(y_test,y_test_predict)]

models_list=list(zip(Models,MAE,MSE,RMSE,R2_score))

print("Testing Dataset Outcomes")
models_testing=pd.DataFrame(models_list,columns=['Models','MAE','MSE','RMSE','R2_score'])
models_testing
```
